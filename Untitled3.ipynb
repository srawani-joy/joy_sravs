```python

app_code = r""" ... """
import os, io, time, textwrap, tempfile, json
import streamlit as st
import fitz  # PyMuPDF
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer
from huggingface_hub import login
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# ---------------------------
# UI: Page config
# ---------------------------
st.set_page_config(page_title="StudyMate – PDF Q&A (Granite RAG)", page_icon="📚", layout="wide")

st.title("📚 StudyMate: PDF-based Q&A (IBM Granite + RAG)")
st.caption("Upload PDFs ➜ build semantic index ➜ ask questions ➜ grounded answers with citations")

with st.sidebar:
    st.header("⚙️ Settings")
    hf_token = st.text_input("Hugging Face Token (read access)", type="password", help="Paste a HF token if needed (or leave blank for public models).")
    default_model = "ibm-granite/granite-3.3-8b-instruct"
    model_id = st.text_input("Granite model id", value=default_model, help="Any compatible causal LM on HF.")
    dtype_choice = st.selectbox("Inference dtype", ["auto", "float16", "bfloat16", "float32"], index=0)
    use_4bit = st.checkbox("Load in 4-bit (bitsandbytes)", value=True, help="Great for Colab GPUs; disables on CPU.")
    top_k = st.slider("Top-k chunks to retrieve", 1, 10, 4)
    max_new_tokens = st.slider("Max new tokens", 64, 2048, 512, step=64)
    temperature = st.slider("Temperature", 0.0, 1.5, 0.2, step=0.05)
    add_citations = st.checkbox("Show citations in answer", value=True)
    st.markdown("---")
    st.write("**Embeddings:** `sentence-transformers/all-MiniLM-L6-v2`")
    st.write("**Indexer:** FAISS (L2 over normalized vectors)")

# ---------------------------
# Auth (optional)
# ---------------------------
if hf_token:
    try:
        login(token=hf_token, add_to_git_credential=False)
        os.environ["HUGGINGFACEHUB_API_TOKEN"] = hf_token
        st.sidebar.success("Hugging Face token set")
    except Exception as e:
        st.sidebar.error(f"HF login failed: {e}")

# ---------------------------
# Utilities
# ---------------------------
def extract_text_from_pdf(file_bytes: bytes) -> str:
    doc = fitz.open(stream=file_bytes, filetype="pdf")
    texts = []
    for page in doc:
        # Use PyMuPDF's layout-aware text
        text = page.get_text("text")
        if text:
            texts.append(text)
    return "\n".join(texts)

def chunk_text(text: str, chunk_size: int = 900, chunk_overlap: int = 150):
    # simple word-based chunking with overlap
    words = text.split()
    chunks = []
    i = 0
    while i < len(words):
        chunk_words = words[i:i+chunk_size]
        chunk = " ".join(chunk_words)
        chunks.append(chunk)
        i += (chunk_size - chunk_overlap)
    return chunks

@st.cache_resource(show_spinner=False)
def get_embedder():
    return SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

def normalize(vecs: np.ndarray) -> np.ndarray:
    norms = np.linalg.norm(vecs, axis=1, keepdims=True) + 1e-12
    return vecs / norms

def build_faiss_index(embeddings: np.ndarray):
    # Cosine sim via L2 on normalized vectors
    d = embeddings.shape[1]
    index = faiss.IndexFlatIP(d)  # inner product on normalized vectors == cosine
    index.add(embeddings.astype(np.float32))
    return index

def device_map_auto():
    if torch.cuda.is_available():
        return "auto"
    return {"": "cpu"}

@st.cache_resource(show_spinner=True)
def load_llm(model_id: str, dtype_choice: str, use_4bit: bool):
    torch_dtype = None
    if dtype_choice == "float16":
        torch_dtype = torch.float16
    elif dtype_choice == "bfloat16":
        torch_dtype = torch.bfloat16
    elif dtype_choice == "float32":
        torch_dtype = torch.float32
    else:
        torch_dtype = None  # let Transformers decide

    quantization_config = None
    load_in_4bit = use_4bit and torch.cuda.is_available()
    model_kwargs = {}
    if load_in_4bit:
        try:
            from transformers import BitsAndBytesConfig
            quantization_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_compute_dtype=torch_dtype or torch.bfloat16)
            model_kwargs["quantization_config"] = quantization_config
            model_kwargs["device_map"] = "auto"
        except Exception as e:
            st.warning(f"4-bit quantization not available: {e}")
            load_in_4bit = False

    if not load_in_4bit:
        model_kwargs["device_map"] = device_map_auto()
        if torch_dtype is not None:
            model_kwargs["torch_dtype"] = torch_dtype

    tok = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True, use_auth_token=bool(os.environ.get("HUGGINGFACEHUB_API_TOKEN")))
    mdl = AutoModelForCausalLM.from_pretrained(model_id, trust_remote_code=True, use_auth_token=bool(os.environ.get("HUGGINGFACEHUB_API_TOKEN")), **model_kwargs)
    return tok, mdl

def format_rag_prompt(question: str, contexts: list[str]) -> str:
    system = (
        "You are StudyMate, a helpful academic assistant. "
        "Answer the user's question using ONLY the provided context. "
        "If the answer isn't in the context, say you don't have enough information. "
        "Be concise and cite sources as [#] that map to the context snippets."
    )
    stitched = ""
    for i, c in enumerate(contexts, 1):
        stitched += f"[{i}] {c}\n\n"
    user = f"Question: {question}\n\nContext:\n{stitched}\nAnswer:"
    return f"<|system|>\n{system}\n<|user|>\n{user}\n<|assistant|>\n"

def generate_answer(tok, mdl, prompt, max_new_tokens=512, temperature=0.2):
    input_ids = tok(prompt, return_tensors="pt").to(mdl.device)
    with torch.no_grad():
        out = mdl.generate(
            **input_ids,
            max_new_tokens=max_new_tokens,
            do_sample=(temperature > 0),
            temperature=max(temperature, 1e-6),
            pad_token_id=tok.eos_token_id
        )
    full = tok.decode(out[0], skip_special_tokens=True)
    # try to split at last assistant marker if present
    if "<|assistant|>" in full:
        full = full.split("<|assistant|>")[-1].strip()
    return full.strip()

# ---------------------------
# App state
# ---------------------------
if "docs" not in st.session_state:
    st.session_state.docs = []   # list of dicts: {name, text, chunks}
if "index" not in st.session_state:
    st.session_state.index = None
if "chunk_map" not in st.session_state:
    st.session_state.chunk_map = []  # (doc_name, chunk_text)
if "emb_model_dim" not in st.session_state:
    st.session_state.emb_model_dim = 384  # all-MiniLM-L6-v2

# ---------------------------
# Upload PDFs and build index
# ---------------------------
st.subheader("1) Upload PDFs")
files = st.file_uploader("Drop one or more PDFs", type=["pdf"], accept_multiple_files=True)

col_build_l, col_build_r = st.columns([3,2])
with col_build_l:
    build_btn = st.button("Build / Rebuild Index", type="primary")
with col_build_r:
    clear_btn = st.button("Clear All", type="secondary")

if clear_btn:
    st.session_state.docs = []
    st.session_state.index = None
    st.session_state.chunk_map = []
    st.success("Cleared in-memory documents & index.")

if build_btn:
    if not files:
        st.warning("Please upload at least one PDF.")
    else:
        with st.spinner("Extracting text and building FAISS index..."):
            docs = []
            all_chunks = []
            chunk_map = []
            for f in files:
                raw = f.read()
                text = extract_text_from_pdf(raw)
                chunks = chunk_text(text)
                docs.append({"name": f.name, "text": text, "chunks": chunks})
                for c in chunks:
                    chunk_map.append((f.name, c))
                    all_chunks.append(c)

            embedder = get_embedder()
            emb = embedder.encode(all_chunks, convert_to_numpy=True, show_progress_bar=True, batch_size=64, normalize_embeddings=True)
            index = build_faiss_index(emb)
            st.session_state.docs = docs
            st.session_state.index = index
            st.session_state.chunk_map = chunk_map
        st.success(f"Indexed {len(all_chunks)} chunks from {len(docs)} PDF(s).")

st.divider()

# ---------------------------
# Q&A
# ---------------------------
st.subheader("2) Ask a Question")
query = st.text_input("Your question", placeholder="e.g., What is the main theorem in chapter 3 and its assumptions?")
ask = st.button("Ask")

if ask:
    if st.session_state.index is None or not st.session_state.chunk_map:
        st.warning("Please upload PDFs and build the index first.")
    elif not query.strip():
        st.warning("Enter a question.")
    else:
        embedder = get_embedder()
        q_emb = embedder.encode([query], convert_to_numpy=True, normalize_embeddings=True)
        D, I = st.session_state.index.search(q_emb.astype(np.float32), top_k)
        I = I[0].tolist()
        retrieved = [st.session_state.chunk_map[i] for i in I]  # list of (doc_name, chunk)
        contexts = [c for (_, c) in retrieved]
        prompt = format_rag_prompt(query, contexts)

        with st.spinner("Loading Granite model (first time may take a bit)…"):
            tok, mdl = load_llm(model_id, dtype_choice, use_4bit)

        with st.spinner("Generating answer…"):
            answer = generate_answer(tok, mdl, prompt, max_new_tokens=max_new_tokens, temperature=temperature)

        # Display
        st.markdown("### ✨ Answer")
        st.write(answer)

        if add_citations:
            st.markdown("### 📎 Sources")
            for i, (doc_name, chunk) in enumerate(retrieved, 1):
                with st.expander(f"[{i}] {doc_name}"):
                    st.write(textwrap.shorten(chunk, width=1200, placeholder=" …"))

st.info("Tip: If you change the model or token, rebuild isn’t required. It only affects generation.")
"""
with open("/content/app.py", "w") as f:
    f.write(app_code)

print("app.py written.")
```

```python
import PyPDF2
from sentence_transformers import SentenceTransformer
import faiss
import openai

# --- Configuration ---
OPENAI_API_KEY = "YOUR_OPENAI_API_KEY"  # Replace with your key
PDF_PATH = "sample.pdf"  # Path to your PDF file

openai.api_key = OPENAI_API_KEY

# --- Step 1: Extract text from PDF ---
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text

# --- Step 2: Split text into chunks ---
def split_text(text, chunk_size=500):
    words = text.split()
    return [' '.join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]

# --- Step 3: Embed chunks ---
def embed_chunks(chunks, model):
    return model.encode(chunks)

# --- Step 4: Build FAISS index ---
def build_faiss_index(embeddings):
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings)
    return index

# --- Step 5: Find relevant context ---
def find_relevant_chunk(question, chunks, model, index, embeddings, top_k=1):
    q_emb = model.encode([question])
    D, I = index.search(q_emb, top_k)
    return [chunks[i] for i in I[0]]

# --- Step 6: Ask LLM ---
def ask_llm(question, context):
    prompt = f"Context:\n{context}\n\nQuestion: {question}\nAnswer:"
    response = openai.Completion.create(
        engine="text-davinci-003",  # Or "gpt-3.5-turbo-instruct"
        prompt=prompt,
        max_tokens=150,
        temperature=0.2,
    )
    return response.choices[0].text.strip()

# --- Main workflow ---
def main():
    # 1. Extract text
    text = extract_text_from_pdf(PDF_PATH)
    # 2. Split into chunks
    chunks = split_text(text)
    # 3. Embed chunks
    model = SentenceTransformer('all-MiniLM-L6-v2')
    embeddings = embed_chunks(chunks, model)
    # 4. Build index
    index = build_faiss_index(embeddings)
    # 5. User Q&A loop
    print("Ask a question about the PDF (type 'exit' to quit):")
    while True:
        question = input("Question: ")
        if question.lower() == "exit":
            break
        relevant_context = find_relevant_chunk(question, chunks, model, index, embeddings)[0]
        answer = ask_llm(question, relevant_context)
        print(f"Answer: {answer}\n")

if __name__ == "__main__":
    main()
```

```python
# 1. Install dependencies (run this cell first)
!pip install transformers sentence-transformers PyPDF2 gradio

# 2. Imports and setup
import gradio as gr
import PyPDF2
from sentence_transformers import SentenceTransformer
from transformers import pipeline
import numpy as np

# 3. PDF text extraction
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text

# 4. Chunking for retrieval
def chunk_text(text, chunk_size=500):
    sentences = text.split('. ')
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        if len(current_chunk) + len(sentence) < chunk_size:
            current_chunk += sentence + ". "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence + ". "
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

# 5. Embedding model
embedder = SentenceTransformer('all-MiniLM-L6-v2')

def embed_chunks(chunks):
    return embedder.encode(chunks)

# 6. Semantic search
def find_most_similar_chunk(query, chunks, chunk_embeddings):
    query_embedding = embedder.encode([query])[0]
    similarities = np.dot(chunk_embeddings, query_embedding)
    most_similar_idx = np.argmax(similarities)
    return chunks[most_similar_idx]

# 7. QA model
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

def answer_question(question, context):
    result = qa_pipeline({'question': question, 'context': context})
    return result['answer']

# 8. The main PDF QA system
def pdf_qa_system(pdf_path, question):
    text = extract_text_from_pdf(pdf_path)
    chunks = chunk_text(text)
    chunk_embeddings = embed_chunks(chunks)
    context = find_most_similar_chunk(question, chunks, chunk_embeddings)
    answer = answer_question(question, context)
    return answer

# 9. Gradio interface
def gradio_qa(pdf_file, question):
    pdf_path = pdf_file.name
    return pdf_qa_system(pdf_path, question)

import gradio as gr

def return_pdf_link(pdf_file):
    # Just return the file object; Gradio will make a temporary download link
    return pdf_file

iface = gr.Interface(
    fn=return_pdf_link,
    inputs=gr.File(label="Upload PDF"),
    outputs=gr.File(label="Download Link"),
    title="Get PDF Download Link"
)

iface.launch()

```
